```python
#Performing Image erosion and dilation serially
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt

# Reading the input image
img = cv2.imread("metal114.jpg")
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Taking a matrix of size 5 as the kernel
kernel = np.ones((3, 3), np.uint8)
 
# The first parameter is the original image kernel is the matrix with which image is convolved and third parameter is the number
# of iterations, which will determine how much you want to erode/dilate a given image.
img_erosion = cv2.erode(img, kernel, iterations=1)
img_dilation = cv2.dilate(img_erosion, kernel, iterations=1)

rgb_img_erosion = cv2.cvtColor(img_erosion, cv2.COLOR_BGR2RGB)
rgb_img_dilation = cv2.cvtColor(img_dilation, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)
```

```python
#After Erosion
plt.imshow(rgb_img_erosion)
```

```python
#Dilated
plt.imshow(rgb_img_dilation)
```

```python
#Applying the Gaussian Blur
gaus_image = cv2.GaussianBlur(rgb_img_dilation, (9, 9),0)
plt.figure(figsize=(11,6))
plt.subplot(121), plt.imshow(rgb_img_dilation),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(gaus_image),plt.title('Gaussian Filter')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
#Applying Median Blur
med_image = cv2.medianBlur(gaus_image, 9)
plt.figure(figsize=(11,6))
plt.subplot(121), plt.imshow(gaus_image),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(med_image),plt.title('Median Filter')
plt.xticks([]), plt.yticks([])
plt.show()
```
